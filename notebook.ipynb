```python
import pandas as pd
data = pd.read_csv("data.csv") 
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
data['tpep_dropoff_datetime'] = pd.to_datetime(data['tpep_dropoff_datetime'])
```

```python
#1 Extract all trips with trip_distance larger than 50.
trips = data[data['trip_distance']>50]
trips.head()
```

```python
#2 Extract all trips where payment_type is missing.
trips_nopayment = data[pd.isna(data['payment_type'])]
trips_nopayment.head()
```

```python
#3 For each (PULocationID, DOLocationID) pair, determine the number of trips.
trip_counts = data.groupby(['PULocationID', 'DOLocationID']).size().reset_index(name='trip_count')
trip_counts.head()
```

```python
#4 Save all rows with missing VendorID, passenger_count, store_and_fwd_flag, or payment_type in a new dataframe called bad, and remove those rows
#from the original dataframe.

missing = data[data[['VendorID', 'passenger_count', 'store_and_fwd_flag', 'payment_type']].isna().any(axis=1)]
data = data.dropna(subset=['VendorID', 'passenger_count', 'store_and_fwd_flag', 'payment_type'])
data.head()
```

```python
#5 Add a duration column storing how long each trip has taken (use tpep_pickup_datetime, tpep_dropoff_datetime).
data['duration']= data['tpep_dropoff_datetime'] - data['tpep_pickup_datetime']
data.head()
```

```python
#6 For each pickup location, determine how many trips have started there.
trips_for_locations = data.groupby('PULocationID').size().reset_index(name='trip_count')
trips_for_locations.head()
```

```python
#7 Cluster the pickup time of the day into 30-minute intervals (e.g., from 02:00 to 02:30).
data['pickup_datetime30m'] = data['tpep_pickup_datetime'].dt.floor('30T').dt.time
data.head()
```

```python
#8 For each interval, determine the average number of passengers and the average fare amount.
avg_passengers = data.groupby('pickup_datetime30m').agg(
    avg_passenger_count=('passenger_count', 'mean'),
    avg_fare_amount=('fare_amount', 'mean')
).reset_index()
avg_passengers.head()
```

```python
#9 For each payment type and each interval, determine the average fare amount.
avg_fare = data.groupby(['payment_type','pickup_datetime30m']).agg(
    avg_fare_amount=('fare_amount', 'mean')
).reset_index()
avg_fare.head()
```

```python
#10 For each payment type, determine the interval when the average fare amount is maximum.
max_fare = avg_fare.loc[avg_fare.groupby('payment_type')['avg_fare_amount'].idxmax()]
max_fare
```

```python
# 11 For each payment type, determine the interval when the overall ratio between the tip and the fare amounts is maximum.
data['tip_fare_ratio'] = data['tip_amount'] / data['fare_amount']
data = data.dropna(subset=['tip_fare_ratio'])
avg_tip_fare = data.groupby(['payment_type', 'pickup_datetime30m']).agg(
    avg_tip_fare_ratio=('tip_fare_ratio', 'mean')
).reset_index()
max_tip_fare = avg_tip_fare.loc[avg_tip_fare.groupby('payment_type')['avg_tip_fare_ratio'].idxmax()]
max_tip_fare
```

```python
#12 Find the location with the highest average fare amount.
location_avg_fare = data.groupby('PULocationID').agg(
    avg_tip_fare=('fare_amount', 'mean')
).reset_index()
max_loc_fare = location_avg_fare.loc[location_avg_fare['avg_tip_fare'].idxmax()]
max_loc_fare
```

```python
# 13 Build a new dataframe (called common) where, for each pickup location, 
# we keep all trips to the 5 most common destinations (i.e., each pickup 
# location can have different common destinations).

trip_counts = data.groupby(['PULocationID', 'DOLocationID']).size().reset_index(name='trip_count')
top_5 = trip_counts.sort_values(['PULocationID', 'trip_count'], ascending=[True, False]) \
                                 .groupby('PULocationID').head(5)
common = data.merge(top_5[['PULocationID', 'DOLocationID']], on=['PULocationID', 'DOLocationID'])
common.head()
```

```python
#14 On the common dataframe, for each payment type and each interval, determine the average fare amount.
avg_fare_common = common.groupby(['payment_type','pickup_datetime30m']).agg(
    avg_fare_amount=('fare_amount', 'mean')
).reset_index()
avg_fare_common.head()
```

```python
#15 Compute the difference of the average fare amount computed in the previous point with those computed in point 9.
diff_fare = avg_fare_common['avg_fare_amount'] - avg_fare['avg_fare_amount']
diff_fare.head()
```

```python
# 16 Compute the ratio between the differences computed in the previous point and those computed in point 9.
#  Note: you have to compute a ratio for each pair (payment type, interval).
# Supponiamo di avere i DataFrames avg_fare_common e avg_fare come definiti nei punti 15 e 9

merged_fares = avg_fare_common.merge(avg_fare, on=['payment_type', 'pickup_datetime30m'], suffixes=('_common', '_tot'))
merged_fares['diff_fare'] = merged_fares['avg_fare_amount_common'] - merged_fares['avg_fare_amount_tot']


merged_fares['ratio'] = merged_fares['diff_fare'] / merged_fares['avg_fare_amount_tot']
merged_fares[['payment_type', 'pickup_datetime30m', 'avg_fare_amount_common', 'avg_fare_amount_tot', 'diff_fare', 'ratio']]
```

```python
# Build chains of trips. Two trips are consecutive in a chain
data = data.sort_values(by=['VendorID', 'tpep_dropoff_datetime', 'tpep_pickup_datetime']).reset_index(drop=True)
data['chain'] = None
chain_id = 0

#check contitions
for i in range(1, len(data)):
    if (data.loc[i, 'VendorID'] == data.loc[i - 1, 'VendorID'] and
        data.loc[i, 'PULocationID'] == data.loc[i - 1, 'DOLocationID'] and
        data.loc[i, 'tpep_pickup_datetime'] > data.loc[i - 1, 'tpep_dropoff_datetime'] and
        (data.loc[i, 'tpep_pickup_datetime'] - data.loc[i - 1, 'tpep_dropoff_datetime']).total_seconds() <= 120):
        
        
        data.loc[i, 'chain'] = chain_id
    else:
        chain_id += 1
        data.loc[i, 'chain'] = chain_id

data.loc[0, 'chain'] = 0
```
